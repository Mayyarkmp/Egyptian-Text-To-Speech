```python
import pandas as pd
from pyarabic.normalize import strip_tashkeel, strip_tatweel, normalize_lamalef
```

```python
df = pd.read_csv('./datasets_csv/converted_audios_to_text.csv')
df
```

```python
normalized_texts = []
texts = df['text'].tolist()

for text in texts:
    normalized_text = strip_tashkeel(text)
    normalized_text = strip_tatweel(normalized_text)
    normalized_text = normalize_lamalef(normalized_text)
    normalized_texts.append(normalized_text)
```

```python
csv_len = len(df)
df['normalized_text'] = normalized_texts

df['is_remove'] = [0] * csv_len
```

```python
df
```

```python
pattern = r'موسيقى]|صوت صوت'
df.loc[df['text'].str.contains(pattern, regex=True), 'is_remove'] = 1
```

```python
df1 = df.iloc[:len(df)//2]
df2 = df.iloc[len(df)//2:]

df1.to_csv('datasets_csv/audio_text_datasets/audio_text_amir.csv', encoding='utf-8-sig', index=False)
df2.to_csv('datasets_csv/audio_text_datasets/audio_text_youssef.csv', encoding='utf-8-sig', index=False)
```
