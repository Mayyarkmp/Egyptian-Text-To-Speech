```python
from pydub import AudioSegment
import matplotlib.pyplot as plt
import numpy as np

# Load the sound file (you can specify any supported format)
audio = AudioSegment.from_file("./audios/Akhdar_audio_0.wav")

audio = audio[:100000]
# Set the frame rate for the visualization (smaller window size gives more detailed graphs)
window_size_ms = 50  # 50ms windows

# Get dBFS values for each segment
db_values = [
    audio[i:i+window_size_ms].dBFS
    for i in range(0, len(audio), window_size_ms)
]

# Time axis for plotting
time_axis = np.arange(0, len(db_values)) * (window_size_ms / 1000.0)  # Convert to seconds

# Plotting the dBFS values
plt.figure(figsize=(10, 4))
plt.plot(time_axis, db_values, label="dBFS")
plt.xlabel('Time (s)')
plt.ylabel('dBFS')
plt.title('dBFS over Time')
plt.grid(True)
plt.show()
```

```python
# Import the AudioSegment class for processing audio and the
# split_on_silence function for separating out silent chunks.
import os
from pydub import AudioSegment
from pydub.silence import split_on_silence
from tqdm import tqdm

# Define a function to normalize a chunk to a target amplitude.
def match_target_amplitude(aChunk, target_dBFS):
    ''' Normalize given audio chunk '''
    change_in_dBFS = target_dBFS - aChunk.dBFS
    return aChunk.apply_gain(change_in_dBFS)

# Load your audio.
# loop through audios with files ending in mp4
for f in os.listdir("audios"):
  if not f.endswith(".mp3"):
    print('skipping', f)
    continue # skip files that are not mp4
  print('processing', f)
  song = AudioSegment.from_file(os.path.join("audios", f), format="mp4")

  # Split track where the silence is 2 seconds or more and get chunks using
  # the imported function.
  chunks = split_on_silence (
      # Use the loaded audio.
      song,
      # Specify that a silent chunk must be at least 2 seconds or 2000 ms long.
      min_silence_len = silence_len,
      # Consider a chunk silent if it's quieter than -16 dBFS.
      # (You may want to adjust this parameter.)
      silence_thresh = silence_thresh
  )

  print('processing', len(chunks))
  # Process each chunk with your parameters
  for i, chunk in tqdm(enumerate(chunks)):
      # Create a silence chunk that's 0.5 seconds (or 500 ms) long for padding.
      silence_chunk = AudioSegment.silent(duration=50)

      # Add the padding chunk to beginning and end of the entire chunk.
      audio_chunk = silence_chunk + chunk + silence_chunk

      # Normalize the entire chunk.
      normalized_chunk = match_target_amplitude(audio_chunk, -20.0)

      # Export the audio chunk with new bitrate.
      #print("Exporting chunk{0}.mp3.".format(i))
      normalized_chunk.export(
          ".//dataset/{f}-chunk{i}.mp3".format(f=f, i=i),
          bitrate = "192k",
          format = "mp3"
      )
```

```python
from pydub import AudioSegment
from pydub.silence import split_on_silence
import os

# Function to normalize a chunk to a target amplitude.
def match_target_amplitude(aChunk, target_dBFS):
    ''' Normalize given audio chunk '''
    change_in_dBFS = target_dBFS - aChunk.dBFS
    return aChunk.apply_gain(change_in_dBFS)

# Function to convert milliseconds to SRT time format
def format_time(ms):
    hours = ms // (1000 * 60 * 60)
    ms %= (1000 * 60 * 60)
    minutes = ms // (1000 * 60)
    ms %= (1000 * 60)
    seconds = ms // 1000
    ms %= 1000
    return f"{hours:02}:{minutes:02}:{seconds:02},{ms:03}"

# Function to convert a VTT file into an SRT format
def convert_vtt_to_srt_chunked(vtt_path):
    subtitles = []
    with open(vtt_path, 'r', encoding='utf-8') as vtt_file:
        subtitle_text = ""
        for line in vtt_file:
            # Skip header and empty lines
            if line.strip() == 'WEBVTT' or line.strip() == '':
                continue
            
            # Convert timestamp line from VTT to SRT format
            if '-->' in line:
                line = line.replace('.', ',')
                subtitles.append({'timestamp': line.strip(), 'text': subtitle_text.strip()})
                subtitle_text = ""
            else:
                subtitle_text += line.strip() + " "
    
    return subtitles

# Path to VTT file
vtt_file_path = "./dataset/audio_subtitles.vtt"  # Update with your actual VTT file path

# Load your audio file
song = AudioSegment.from_mp3("./dataset/audio1.mp4-chunk0.mp3")

# Split track where the silence is 2 seconds or more and get chunks
chunks = split_on_silence(
    song,
    min_silence_len=750,
    silence_thresh=-60
)

print('Processing', len(chunks))

# Check if VTT file exists and convert to subtitles for SRT
if os.path.exists(vtt_file_path):
    subtitles = convert_vtt_to_srt_chunked(vtt_file_path)
else:
    print("VTT file not found, proceeding without subtitle text.")
    subtitles = None

# Initialize total time tracker
current_time = 0

# Open SRT file for writing
srt_filename = "./dataset_2/output.srt"
with open(srt_filename, 'w', encoding='utf-8') as srt_file:
    # Process each chunk
    for i, chunk in enumerate(chunks):
        # Create a silence chunk for padding
        silence_chunk = AudioSegment.silent(duration=50)
        audio_chunk = silence_chunk + chunk + silence_chunk

        # Normalize the chunk
        normalized_chunk = match_target_amplitude(audio_chunk, -20.0)

        # Define start and end time for SRT
        chunk_start_time = current_time
        chunk_end_time = current_time + len(audio_chunk)

        # Write SRT entry for this chunk
        srt_file.write(f"{i + 1}\n")
        srt_file.write(f"{format_time(chunk_start_time)} --> {format_time(chunk_end_time)}\n")

        # Write subtitle text if available from VTT, otherwise leave it empty
        if subtitles and i < len(subtitles):
            srt_file.write(subtitles[i]['text'] + "\n")
        else:
            srt_file.write("[Audio without subtitle text]\n")  # Placeholder if no text

        srt_file.write("\n")

        # Update the current time for the next chunk
        current_time = chunk_end_time

        # Export the chunk
        chunk_filename = f"./dataset/chunk{i}.mp3"
        print(f"Exporting {chunk_filename}")
        normalized_chunk.export(chunk_filename, bitrate="192k", format="mp3")

print(f"SRT file written to {srt_filename}")
```
