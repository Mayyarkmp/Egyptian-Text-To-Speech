```python
def remove_audio(filename):
    audios_to_remove = ['akhdar_audio_173', 'akhdar_audio_172', 'akhdar_audio_171', 'akhdar_audio_170', 
                    'akhdar_audio_160', 'akhdar_audio_164', 'akhdar_audio_169', 'akhdar_audio_168', 
                    'akhdar_audio_14', 'akhdar_audio_18', 'akhdar_audio_28', 'akhdar_audio_162', 
                    'akhdar_audio_165', 'akhdar_audio_166', 'akhdar_audio_167', 'akhdar_audio_32', 
                    'akhdar_audio_35', 'akhdar_audio_37', 'akhdar_audio_39', 'akhdar_audio_98', 'akhdar_audio_118']

    for audio_to_remove in audios_to_remove:
        if audio_to_remove in filename:
            return True

    return False    
```

```python
df = df[~df['filename'].apply(lambda x: remove_audio(x))]
```

```python
df.to_csv('./datasets_csv/akhdar_data_cleaned_final.csv', encoding='utf-8-sig', index=False)
```

```python
import os
import librosa
import soundfile as sf

filenames = df['filename'].tolist()

def convert_audios_sr(filenames, source_path, distination_path, target_sr=22050):

    for i, audio_name in enumerate(filenames):
        audio_path = os.path.join(source_path, audio_name)
        audio_final_path = os.path.join(distination_path, audio_name)
        audio, sr = librosa.load(audio_path, sr=None)
        resampled_audio = librosa.resample(audio, orig_sr=sr, target_sr=target_sr)
        sf.write(audio_final_path, resampled_audio, target_sr)
        print(f'Converted audio {i}: {audio_name}')



source_path = './audios/wavs_old'
distination_path = './audios/wavs'
convert_audios_sr(filenames, source_path, distination_path)
```

```python
import librosa
import os
from collections import Counter

# Directory containing audio files
audio_dir = "./audios/wavs"

# Collect sample rates
sample_rates = []

for i, file in enumerate(os.listdir(audio_dir)):
    if file.endswith(".wav"):  # Adjust the extension if necessary
        print(f'Processed audio {i}')
        filepath = os.path.join(audio_dir, file)
        _, sr = librosa.load(filepath, sr=None)  # sr=None preserves original sample rate
        sample_rates.append(sr)

# Analyze sample rate distribution
counter = Counter(sample_rates)
print("Sample Rate Distribution:")
for rate, count in counter.items():
    print(f"{rate} Hz: {count} files")

# Suggest the most common sample rate
most_common_rate = counter.most_common(1)[0][0]
print(f"Suggested Sample Rate: {most_common_rate} Hz")
```

### Move audios

```python
import os
import shutil
import pandas as pd

def copy_audios(source, distination, csv_path):
    df = pd.read_csv(csv_path)
    filenames = df['filename'].tolist()
    os.makedirs(distination, exist_ok=True)

    for filename in filenames:
        filename = filename + '.wav'
        filename_source_path = os.path.join(source, filename)
        filename_distination_path = os.path.join(distination, filename)
        shutil.copy2(filename_source_path, filename_distination_path)


def compare(csv_path, folder_path):
    filenames_folder = os.listdir(folder_path)
    df = pd.read_csv(csv_path)
    filenames_csv = df['filename'].tolist()

    filenames_csv = [filename + '.wav' for filename in filenames_csv]

    filenames_folder.sort()
    filenames_csv.sort()

    if filenames_csv == filenames_folder:
        print('They are the same')
    else:
        print('They are different')


if __name__ == '__main__':
    # source_folder = ''
    # distination_folder = ''
    # csv_path = ''
    # copy_audios(source_folder, distination_folder, csv_path)
    compare(csv_path='', folder_path='')
```
