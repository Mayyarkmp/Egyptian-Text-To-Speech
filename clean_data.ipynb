## Clean Akhdar Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('./datasets_csv/main_datasets/akhdar_data.csv')
df
```

```python
index = df[df['video_title'] == 'ستيف جوبز - مراجعة كرتونية ل كتاب والتر إيزاكسون'].index.values[0]
df = df.iloc[1:index+1]
```

```python
index = df[df['video_title'] == 'كتاب قلق السعي إلى المكانة  🔥 #أخضر'].index.values[0]

df.loc[index:, 'voice'] = 'voice 2'
df
```

```python
df = df[(df['video_title'] != 'أخضر بدأ بيكم من 5 سنين ووصل لمليون شخص في 173 دولة في العالم 🗺️ #أخضر #مليون_أخضر') & (df['video_title'] != 'أخضر بدأ المشي')]
df
```

```python
sum1 = df['video_duration (min)'].sum()
```

```python
df.to_csv('./datasets_csv/cleaned_datasets/akhdar_data_cleaned.csv', encoding='utf-8-sig', index=False)
```

## Clean ReadTube Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('./datasets_csv/main_datasets/readtube_data.csv')
df
```

```python
index = df[df['video_title'] == 'كيف تؤسس مشروعا ناجحا  🚀 ملخص كتاب : رائد الأعمال'].index.values[0]
df = df.iloc[1:index+1]
```

```python
df = df[~df['video_title'].str.contains("Q&A", case=False, na=False)]
df
```

```python
sum2 = df['video_duration (min)'].sum()
```

```python
df.to_csv('./datasets_csv/cleaned_datasets/readtube_cleaned.csv', encoding='utf-8-sig', index=False)
```

## Clean Ali Muhammad Ali Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('./datasets_csv/main_datasets/ali_muhammad_ali_data.csv')
df
```

```python
df = df[df['video_duration (min)'] > 4]
df
```

```python
df = df[~df['video_title'].str.contains("بودكاست إطناب", case=False, na=False)]
df
```

```python
df = df[(df['video_title'] != 'مهارة صناعة الفيديو باستخدام الهاتف الذكي') & (df['video_title'] != 'غزة وآلام الخذلان')]
df.iloc[0, df.columns.get_loc('video_duration (min)')] = 0
df
```

```python
sum3 = df['video_duration (min)'].sum()
```

```python
df.to_csv('./datasets_csv/cleaned_datasets/ali_muhammad_ali_cleaned.csv', encoding='utf-8-sig', index=False)
```

```python
total = sum1 + sum2 + sum3
total
```
